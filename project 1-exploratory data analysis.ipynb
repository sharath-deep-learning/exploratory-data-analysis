```python
# importing all the dependecies

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
pd.pandas.set_option('display.max_columns',None)
```

```python
# reading thr dataset

dataset=pd.read_csv("train.csv")
dataset.head()
```

```python
dataset.shape
```

```python
# create variable which consists of nan values

features_with_na=[features for features in dataset.columns if dataset[features].isnull().sum()>1]

for feature in features_with_na:
    print(feature,np.round(dataset[feature].isnull().mean(),4),'% missing values')
```

```python
# plot the graph of dependent variable vs independent variables

for feature in features_with_na:
    data=dataset.copy()
    
    data[feature]=np.where(data[feature].isnull(),1,0)
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

```python


print("Id of Houses {}".format(len(dataset.Id)))

```

```python
# numerical variables

numerical_features = [feature for feature in dataset.columns if dataset[feature].dtypes != 'O']
print('number of numerical_features',len(numerical_features))
dataset[numerical_features].head()
```

```python
# temporal variables eg.(date /time  variables)

year_feature=[feature for feature in numerical_features if 'Yr'in feature or 'Year' in feature]

year_feature
```

```python


for feature in year_feature:
    print(feature,dataset[feature].unique())
```

```python
# plot the graph b/w independent variavble and temporal variables

dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year sold')
plt.ylabel('Median House Price')
plt.title("House Price vs YearSold")
```

```python
year_feature
```

```python
for feature in year_feature:
    if feature!='YrSold':
        data=dataset.copy()
        data[feature]=data['YrSold']-data[feature]
        
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

```python
# discrete variables

discrete_feature=[feature for feature in numerical_features if len(dataset[feature].unique())<25 and feature not in year_feature+['Id']]
print("discrete_feature count: {}".format(len(discrete_feature)))
```

```python
discrete_feature
```

```python
dataset[discrete_feature].head()
```

```python
# plot thr grapgh b/w independent variables vs discrete variables


for feature in discrete_feature:
    data=dataset.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```

```python
# in discrete variables two types continuos and categorical variables
# lets create variable which holds continuous variables


continous_feature=[feature for feature in numerical_features if feature not in discrete_feature+year_feature+['Id']]
print("continous_feature count {}".format(len(continous_feature)))
```

```python


for feature in continous_feature:
    data=dataset.copy()
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel("count")
    plt.title(feature)
    plt.show()
```

```python
for feature in continous_feature:
    data=dataset.copy()
    
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data['SalePrice']=np.log(data['SalePrice'])
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel("SalePrice")
        plt.title(feature)
        plt.show()
    
```

```python
for feature in continous_feature:
    data=dataset.copy()
    
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
```
